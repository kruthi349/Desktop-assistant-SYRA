```python
from comtypes.client import CreateObject
from comtypes import CLSCTX_ALL
from pycaw.pycaw import IAudioEndpointVolume
import comtypes

# GUIDs manually defined
CLSID_MMDeviceEnumerator = comtypes.GUID("{BCDE0395-E52F-467C-8E3D-C4579291692E}")
IID_IMMDeviceEnumerator = comtypes.GUID("{A95664D2-9614-4F35-A746-DE8DB63617E6}")

try:
    # Create the enumerator
    enumerator = CreateObject(CLSID_MMDeviceEnumerator, interface=comtypes.IUnknown)

    # Query IMMDeviceEnumerator
    enum_iface = enumerator.QueryInterface(
        comtypes.GUID("{A95664D2-9614-4F35-A746-DE8DB63617E6}")
    )

    # Get default audio endpoint
    device = enum_iface.GetDefaultAudioEndpoint(0, 1)  # eRender, Multimedia

    # Activate endpoint volume interface
    interface = device.Activate(
        IAudioEndpointVolume._iid_, CLSCTX_ALL, None
    )
    volume = interface.QueryInterface(IAudioEndpointVolume)

    print("SUCCESS! System endpoint volume active!")
    print("Current Volume:", int(volume.GetMasterVolumeLevelScalar() * 100))

except Exception as e:
    print("FAIL:", e)
```
